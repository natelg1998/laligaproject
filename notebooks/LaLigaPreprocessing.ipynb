```python
# List Imports 
import numpy as np
import pandas as pd
from datetime import datetime as dt
import os
import glob
from pathlib import Path
```

```python
DATA = os.path.realpath('../data/football_uk_data/')
print(DATA)

file_path = Path(os.path.realpath('../data/football_uk_data/'))
```

```python
files = []
for file in os.listdir(DATA):
    files.append(os.path.join(file_path, file))
```

```python
match_data = pd.concat([pd.read_csv(file) for file in files],axis = 0).reset_index(drop=True)
```
